```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
data = pd.read_csv('data6.csv')
data.head()
```

```python
data.columns = ['Date', 'mood','arousal','valence','activity','screentime','other','travel','buu','coso','enga','fiof','teca']
data.head()
```

```python
data.info()
```

```python
data.Date = pd.to_datetime(data.Date)
data.set_index('Date', inplace=True)
```

```python
data.head()
```

```python
data.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Date', fontsize=20)
```

```python
data[['mood']].plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Date', fontsize=20)
```

```python
mood = data[['mood']]
mood.rolling(3).mean().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Date', fontsize=20);
```

```python
arousal = data[['arousal']]
arousal.rolling(3).mean().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Date', fontsize=20);
```

```python
data_rm = pd.concat([mood.rolling(3).mean(), arousal.rolling(3).mean()], axis=1)
data_rm.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Date', fontsize=20);
```

```python
mood.diff().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Date', fontsize=20)
```

```python
sns.lmplot(x='mood', y='arousal', fit_reg=False, data=data)
```

```python
sns.lmplot(x='arousal', y='valence', fit_reg=False, data=data, hue='mood')
```
